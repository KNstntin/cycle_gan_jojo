```
import torch
import torchvision
from torchvision import transforms
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
```

```
from matplotlib.pyplot import imshow
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize(300),
    transforms.RandomCrop(256),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
    transforms.RandomHorizontalFlip(),
])

dataset_horses= dataset.FolderDataSet("./horse2zebra/train/trainA", transform=transform)
dataloader_horses = torch.utils.data.DataLoader(dataset_horses, batch_size = 1, shuffle=True)
imshow(torchvision.transforms.ToPILImage()(dataset.de_norm(dataset_horses[0])))
```

```
dataset_zebras= dataset.FolderDataSet("./horse2zebra/train/trainB", transform=transform)
dataloader_zebras = torch.utils.data.DataLoader(dataset_zebras, batch_size = 1, shuffle=True)
imshow(torchvision.transforms.ToPILImage()(dataset.de_norm(dataset_zebras[0])))
```

```
dataset_test = dataset.FolderDataSet("./horse2zebra/test/testA", transform=transform)
dataloader_test = torch.utils.data.DataLoader(dataset_test, batch_size = 1, shuffle=True)
# imshow(torchvision.transforms.ToPILImage()(torch.clamp(inv_normalize(dataset_test[0]), min = 0, max = 255)))
imshow(torchvision.transforms.ToPILImage()(dataset.de_norm(dataset_test[0])))
```

```
from itertools import chain
device = torch.device("cuda:0") if torch.cuda.is_available() else torch.device("cpu")
pool_images = models.Buffer(device, 50)
pool_targets = models.Buffer(device, 50)
gan = models.CycleGAN(lambd = 10, resnet_layers = 6, device = device, buffer_images=pool_images, dim_gen = 32, buffer_targets=pool_targets)
gan.to(device)
opt_descriminator = torch.optim.Adam(params = chain(gan.gan_straight.descriminator.parameters(), gan.gan_inverse.descriminator.parameters()), lr = 2e-4, betas=(0.5, 0.999))
opt_generator = torch.optim.Adam(params = chain(gan.gan_straight.generator.parameters(), gan.gan_inverse.generator.parameters()), lr = 2e-4, betas=(0.5, 0.999))

# checkpoint = torch.load('/content/drive/MyDrive/Deep learning school/HW6/horse2zebra_weights')
# gan.load_state_dict(checkpoint['gan_state_dict'])
# opt_descriminator.load_state_dict(checkpoint['opt_d_state_dict'])
# opt_generator.load_state_dict(checkpoint['opt_g_state_dict'])
```

```
train.train(gan, dataloader_horses, dataloader_zebras, dataloader_test, opt_generator, opt_descriminator, device, epochs=10)
torch.save({
            'gan_state_dict': gan.state_dict(),
            'opt_d_state_dict': opt_descriminator.state_dict(),
            'opt_g_state_dict': opt_generator.state_dict()
            }, '/content/drive/MyDrive/Deep learning school/HW6/horse2zebra_weights')
```
